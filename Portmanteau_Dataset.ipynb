```python

from bs4 import BeautifulSoup
```

```python
import requests

URL = "https://en.wikipedia.org/wiki/List_of_portmanteaus"
page = requests.get(URL)

#print(page.text)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
results = soup.find_all("div", class_="div-col")
a=results[0].find_all("li")
```

```python
words={}
for i in results:
    tmp = i.find_all("li")
    for j in tmp:
        words[j.find('a').text]=j.text
```

```python
words
```

```python
w1=[]
w2=[]
toDel=[]
k=0
```

```python
for i in words:
    try:
        tmp=words[i].split('and')
        tmp2=tmp[0].split('from')[1]
        w2.append(tmp[1])
        w1.append(tmp2)
    except:
        print("error")
        toDel.append(i)
```

```python
len(words)
```

```python
for i in range(len(w1)):
    w1[i]=w1[i].strip()
    w2[i]=w2[i].strip().split(" ")[0]
```

```python
for i in toDel:
    del words[i]
```

```python
import pandas as pd

df = pd.DataFrame()
```

```python
df['word']=list(words.keys())
df['opposite']=w1
df['rhyme']=w2
```
